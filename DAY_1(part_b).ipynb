```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Initialize encoded_data from df and apply previous transformations
encoded_data = df.copy()

# Apply Label Encoding for 'Sex'
label_encoder = LabelEncoder()
encoded_data['Sex_Label'] = label_encoder.fit_transform(encoded_data['Sex'])

# Apply One-Hot Encoding for 'Embarked'
one_hot_encoding = pd.get_dummies(encoded_data['Embarked'], prefix='Embarked')
encoded_data = pd.concat([encoded_data, one_hot_encoding], axis=1)

#Since 'Pclass' is ordinal, we can create a mapped column for demo
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third'}
#Ordinal Encoding for p class
encoded_data['Pclass_Category'] = encoded_data['Pclass'].map(pclass_mapping)
```

```python
print(encoded_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
numerical_columns = ['Age','Fare','Pclass']

encoded_data.dropna(subset = numerical_columns,inplace=True)
```

```python
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
encoded_data['Age_MinMax'] = minmax_scaler.fit_transform(encoded_data[['Age']])
encoded_data['Fare_MinMax'] = minmax_scaler.fit_transform(encoded_data[['Fare']])
```

```python
encoded_data['Age_Standard'] = standard_scaler.fit_transform(encoded_data[['Age']])
encoded_data['Fare_Standard'] = standard_scaler.fit_transform(encoded_data[['Fare']])
```
